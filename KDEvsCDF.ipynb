```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_palette('tab10')
sns.set_style("whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
norm_arr = np.random.normal(loc=0,scale=2,size=1000)
avg = sum(norm_arr)/len(norm_arr)
std = np.std(norm_arr)
avg,std
```

```python
rand_arr = np.random.normal(loc=2,scale=1,size=2000)
avg = sum(rand_arr)/len(rand_arr)
std = np.std(rand_arr)
avg,std
```

Generate 2 different distribution sample data for demonstrate

# KDE(Kernel Desity Estimator)
## Probability Distribution Estimation using KDE


```python
sns.histplot(norm_arr,kde=True,label='norm')
sns.histplot(rand_arr,kde=True,label='rand')
plt.legend()
```

# CDF(Culmulative Distribution Function)


```python
c_ax = sns.histplot(norm_arr,cumulative=True,color='gray',alpha=0.3)
c_ax.set_ylabel('Cumulative count',color='gray')

c_ax2 = c_ax.twinx()
c_ax2 = sns.kdeplot(norm_arr,cumulative=True,color='r',label='CDF')
c_ax2.set_ylabel('Cumulative probability',color='r')
c_ax2.legend()
```

# Comparing KDE vs CDF

Create 3 different distribution for demonstration

```python
d1 = np.random.normal(-7,1.5,700)
d2 = np.random.normal(0,1,500)
d3 = np.random.normal(6,2,1000)

mulvar_arr = np.concatenate((d1,d2,d3))
```

```python
kde_ax = sns.histplot(
    mulvar_arr,stat='probability',kde=True,kde_kws=dict(bw_adjust=0.6),
    label = 'Probability'
)
kde_ax.set_yticks(np.arange(0,0.14,0.02))

cdf_ax = kde_ax.twinx()
cdf_ax = sns.kdeplot(
    mulvar_arr,cumulative=True,color='r',label = 'Culmulative Probability',
)
cdf_ax.set_ylabel('Cumulative probability',color='r')
cdf_ax.grid(color='red',linestyle = '--',alpha=0.3)
cdf_ax.set_yticks(np.arange(0,1.1,0.1))
```

KDE --> better to visual if it has different distribution of not
CDF --> are harder to distinguish the range of each different distribution but easier to visualize range of overall distribution especially in smaller size of data
